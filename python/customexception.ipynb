Custom exception

```python
class Error(Exception):
    pass

class dobException(Error):
    pass
```

```python
year = int(input("Enter the dob"))
age = 2026 - year

try:
    if age <= 30 and age >= 20:
        print(f'your age {age} is valid and you can apply for the exam')
    else:
        raise dobException
except dobException:
    print(f'your age {age} is not valid to apply for the exam. It must be between 20 and 30')
```
